```python
import pandas as pd
import numpy as np
import requests
import pickle
```

```python
df2=pd.read_excel("crime.xls", skiprows=3, nrows=30)
df2
```

```python
df3=pd.read_json("markers.json")
df3
```

```python
df3[["lat", "long"]]=pd.DataFrame(zip(*df3.position)).T
df3
```

```python
json_req= requests.get("https://jsonplaceholder.typicode.com/users").json()
json_req
```

```python
df4= pd.json_normalize(json_req)
df4
```

## Escribir datos en documentos

```python
for i,df in enumerate([df2,df3,df4]):
    df.to_csv(f"./data/{i}.csv")
    df.to_excel(f"./data/{i}.xlsx")
    df.to_json(f"./data/{i}.json")
```

```python
for i,df in enumerate([df2,df3,df4]):
    with open(f"./data/{i}.pickle", 'wb') as f:
        pickle.dump(df,f,protocol=5)
```

```python
for i,df in enumerate([df2,df3,df4]):
    with open(f"./data/{i}.pickle", 'wb') as f:
        print(pickle.load(f).shape)
```
