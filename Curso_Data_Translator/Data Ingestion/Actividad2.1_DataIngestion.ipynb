```python
dfTodos= requests.get("https://jsonplaceholder.typicode.com/todos").json()
dfTodos
```

```python
dfPhotos= requests.get("https://jsonplaceholder.typicode.com/photos").json()
dfPhotos
```

```python
dfAlbums= requests.get("https://jsonplaceholder.typicode.com/albums").json()
dfAlbums
```

```python
dfTodosN= pd.json_normalize(dfTodos)
dfTodosN.head()
```

```python
dfPhotosN= pd.json_normalize(dfPhotos)
dfPhotosN.head()
```

```python
dfAlbumsN= pd.json_normalize(dfAlbums)
dfAlbumsN.head()
```

```python
for i,df in enumerate([df1,df2,df3,SampleData,DataFood,DataInsurance,dfTodosN,dfPhotosN,dfAlbumsN]):
    df.to_csv(f"./Archivos/{i}.csv")
    df.to_excel(f"./Archivos/{i}.xlsx")
    df.to_json(f"./Archivos/{i}.json")
```

```python
for i, df in enumerate([df1,df2,df3,SampleData,DataFood,DataInsurance,dfTodosN,dfPhotosN,dfAlbumsN]):
    with open(f"./Archivos/{i}.pickle", 'wb') as f:
        pickle.dump(df,f,protocol=5)
```

```python
for i,df in enumerate([df1,df2,df3,SampleData,DataFood,DataInsurance,dfTodosN,dfPhotosN,dfAlbumsN]):
    with open(f"./Archivos/{i}.pickle", 'rb') as f:
        print(pickle.load(f).shape)
```
